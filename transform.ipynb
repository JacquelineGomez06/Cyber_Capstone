```python
import csv
import pandas as pd 

df = pd.read_csv('data/df_encoded.csv')

df
```

```python
Type_df = df.iloc[:, [8,9,10,11,12,13]]

Type_df.reset_index(level = 0, inplace = True)
Type_df = Type_df.rename(columns = {'index':'Type_ID'})
```

```python
Type_df
```

```python
Location_df = df.iloc[:, [14,15,16,17,18,19]]
Location_df
```

```python
Location_df.reset_index(level = 0, inplace = True)
Location_df = Location_df.rename(columns = {'index':'Location_ID'})

Location_df
```

```python
df
```

```python
main_df = df.iloc[:, [0,1,2,3,4,5,6,7]]
```

```python
main_df = main_df.rename(columns = {'Unnamed: 0':'Intrusion_ID'})

main_df
```

```python
# Adding Location_ID from Location_df to create foreign key in PGAdmin

main_df = pd.concat([main_df, Location_df["Location_ID"], Type_df['Type_ID']], axis = 1)

main_df
```

```python
# Adding Intrustion_ID from main_df to create primary key in PGAdmin

Location_df = pd.concat([main_df['Intrusion_ID'], Location_df, Type_df['Type_ID']], axis = 1)

Location_df
```

```python
# Adding Intrusion_ID from main_df to create foregin key in PGAdmin

Type_df = pd.concat([main_df['Intrusion_ID'], Type_df, Location_df['Location_ID']], axis = 1)

Type_df
```

```python
main_df.to_csv('data/main_df.csv', index = False)
```

```python
Type_df.to_csv('data/Type_df.csv', index = False)
```

```python
Location_df.to_csv('data/Location_df.csv', index = False)
```
