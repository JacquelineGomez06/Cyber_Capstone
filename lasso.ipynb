```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python

data=pd.read_csv("df_encoded.csv")
```

```python
data_dc = pd.get_dummies(data, columns=['State'])
data_dc
```

```python
data_dc.columns
```

```python
xnumerical=data_dc.drop(['Unnamed: 0','Name_of_Covered_Entity','Date_of_Breach','Date_Posted_or_Updated','breach_start','year','Individuals_Affected'],axis=1)
listnumerical=xnumerical.columns
print(listnumerical)
```

```python
lasso = Lasso(alpha=1.0)
```

```python
y = data_dc["Individuals_Affected"]
xtrain,xtest,ytrain,ytest,=train_test_split(xnumerical,y, test_size=0.3, random_state=10)
lasso.fit(xtrain, ytrain)
```

```python
ypred = lasso.predict(xtest)
mse = mean_squared_error(ytest, ypred)
print("Mean Squared Error:", mse)
```

```python
coefficients = lasso.coef_

for i, name in enumerate(xtrain.columns):
    print(name, end=": ")
    print(round(coefficients[i], 10))
```
